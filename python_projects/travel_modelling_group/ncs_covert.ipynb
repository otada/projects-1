```python
import csv
from contextlib import contextmanager

def transit_region_code(transit_lines_file):
    region_code = []
    line_code_dict = create_dict_of_old_and_new_line_codes()
    old_line_code = line_code_dict.keys()
    for line in transit_lines_file:
        i = 0
        code_string = ""
        while(i < len(line[0])):
            code = line[0][i]
            if not code.isdigit():
                code_string += code
            else:
                break    
            i+=1
        if code_string in old_line_code:
            code_string_update = line_code_dict[code_string]
            region_code.append(code_string_update)
        else:
            region_code.append(code_string)
    return region_code
  
def transit_route_number(transit_lines_file):
    route_number = []
    for line in transit_lines_file:
        k = 0
        number_string = ""
        while(k < len(line[0])):
            code = line[0][k]
            if code.isdigit():
                number_string += code     
            k+=1
        if len(number_string) < 3:
            number_string = "0" + number_string
        route_number.append(number_string)
    return route_number

def transit_branch_line_letters(transit_lines_file):
    branch_line_letters = []
    for line in transit_lines_file:
        j = 0
        letter_string = ""
        line = line[0][::-1]
        while(j < len(line)):
            code = line[j]
            if not code.isdigit():
                letter_string += code
            else:
                break    
            j+=1
        branch_line_letters.append(letter_string[::-1])
    return branch_line_letters
    
def create_dict_of_old_and_new_line_codes():
    line_code_dict = {}
    with open("line_codes.csv", mode="r") as line_codes:
        line_code_file = csv.reader(line_codes)
        next(line_code_file)
        for line_code in line_code_file:
            old_line_code = line_code[1]
            new_line_code = line_code[2]
            
            if old_line_code != "":
                line_code_dict[old_line_code] = new_line_code
        return line_code_dict

@contextmanager
def open_csv_reader(file_path):
    csv_file = open(file_path, mode="r")
    file = csv.reader(csv_file)
    next(file)
    try:
        yield file, csv_file
    finally:
        csv_file.close()

with open_csv_reader("transit_lines.csv") as file:
    transit_lines_file = file[0]
    csv_file = file[1]
    region_code_list = transit_region_code(transit_lines_file)
    csv_file.seek(0)
    route_number_list = transit_route_number(transit_lines_file)
    csv_file.seek(0)
    branch_line_letters_list = transit_branch_line_letters(transit_lines_file)
    # if int(len(region_code_list)) != int(len(route_number_list)) or int(len(route_number_list)) != int(len(branch_line_letters_list)):
    #     raise Exception ("something is wrong..with your csv file!")
    new_line_code_list = [ region_code + route_number + branch_line_letters  
                          for region_code, route_number, branch_line_letters 
                          in zip(region_code_list, route_number_list, branch_line_letters_list)]
    new_transit_line_dict = {}
    with open("csv_for_amit.csv", mode="w") as filename:
        for i in new_line_code_list:
            print(i, file=filename)
   
```
